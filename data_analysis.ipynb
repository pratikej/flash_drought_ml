```python
import os
from PIL import Image
import torch
import torchvision.transforms.functional as tf
import matplotlib.pyplot as plt
```

```python
base = "data/"
input_folders = ["rainfall/", "mean_temp/", "soil_moisture/", "runoff/", "vp/", "wind_speed/"]
output_folder = "FD/"
```

```python
for input_folder in input_folders:
    files = os.listdir(base + input_folder)
    print("Number of files in", input_folder, ":", len(files))

output_files = os.listdir(base + output_folder)
print("Number of files in", output_folder, ":", len(output_files))
```

```python
# Index to get Information on
index = 4000
```

```python
file_path =  base + output_folder + output_files[index]
im = Image.open(file_path)

arr = tf.to_tensor(im)
print(arr.shape)
print(torch.unique(arr, return_counts=True))
arr[arr < -1e30] = -1
arr[arr == 1] = 0
print(torch.unique(arr, return_counts=True))
plt.imshow(arr[0])
```

```python
def display_input_details(input_folder):
    print(input_folder)

    files = os.listdir(base + input_folder)
    file_path =  base + input_folder + files[index + 67] # Previous day input features to output index
    im = Image.open(file_path)

    arr = tf.to_tensor(im)
    # print(arr.shape)
    print(torch.unique(arr, return_counts=True))
    arr[arr < -1e30] = -1
    # print(torch.unique(arr, return_counts=True))
    plt.imshow(arr[0])
```
